### First Web Scraping Project With Sample Website

#### Importing Necessary Libraries

```python
from bs4 import BeautifulSoup
import requests

```

#### Adjusting URL's

```python

base_url="http://books.toscrape.com/"

part="http://books.toscrape.com/catalogue/page-"



urls=[]
urls.append(base_url)
for i in range(2,51):
    new_url=part+str(i)+".html"
    urls.append(new_url)
```

#### Iterating Over Each Page and Each Item in Pages

```python
for url in urls:
    html_text=requests.get(url)
    soup=BeautifulSoup(html_text.text,"lxml")
    books=soup.find_all("li",class_="col-xs-6 col-sm-4 col-md-3 col-lg-3")

     
    for index,book in enumerate(books):
        
        
        name=book.find("article",class_="product_pod").h3.text
        price=book.find("p",class_="price_color").text[1:]
        av=book.find("p",class_="instock availability").text.strip()
        


        print(name+","+price+","+av)

```

#### Writing Items in Txt File 

```python
with open("output.txt","w",encoding="utf-8") as f:
    for url in urls:
        html_text=requests.get(url)
        soup=BeautifulSoup(html_text.text,"lxml")
        books=soup.find_all("li",class_="col-xs-6 col-sm-4 col-md-3 col-lg-3")

    
        for index,book in enumerate(books):
            name=book.find("article",class_="product_pod").h3.text
            price=book.find("p",class_="price_color").text[1:]
            av=book.find("p",class_="instock availability").text.strip()  



           

            f.write(name+",")
            f.write(price+",")
            f.write(av)
            f.write("\n")
```
